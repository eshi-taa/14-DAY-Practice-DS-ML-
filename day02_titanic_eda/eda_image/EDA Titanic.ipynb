Now fill the missing age data by imputation i.e., by filling in the mean age of all passengers.

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',data=df,palette='winter')
```

We use these average values to impute based on Pclass for age

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

```python
df['Age'] = df[['Age','Pclass']].apply(impute_age,axis=1)
```
